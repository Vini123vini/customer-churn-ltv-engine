# Business Insights

## Key Findings

1. The dataset contains 7043 customer records and 33 features.

2. The churn rate is lower than the non-churn rate, indicating that most customers stay with the company.

3. Customers with Month-to-Month contracts have the highest churn count compared to One-Year and Two-Year contracts.

4. Customers with Two-Year contracts have the lowest churn rate, suggesting that long-term contracts improve customer retention.

5. Male and Female customers show very similar churn behavior, indicating that gender has minimal impact on customer churn.

6. The tenure distribution shows a large number of customers with very low tenure and another concentration of customers with high tenure.

7. Customers with shorter tenure are more likely to leave the company, while long-term customers tend to remain loyal.

## Business Recommendations

- Encourage customers to switch from Month-to-Month plans to One-Year or Two-Year contracts.
- Focus retention campaigns on new customers during their first few months.
- Offer loyalty benefits and discounts to long-term customers.
- Since gender has little effect on churn, retention strategies should focus on contract type and customer tenure instead.
