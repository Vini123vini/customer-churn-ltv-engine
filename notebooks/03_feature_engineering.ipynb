```python
import pandas as pd

df = pd.read_csv("../data/raw/telco_churn.csv")

print(df.shape)
```

```python
df["Total Charges"] = pd.to_numeric(df["Total Charges"], errors="coerce")

df["Avg_Revenue_Per_Month"] = (
    df["Total Charges"] /
    (df["Tenure Months"] + 1)
)
```

```python
df["High_Value_Customer"] = (
    df["Monthly Charges"] > 70
).astype(int)
```

```python
df["Long_Term_Customer"] = (
    df["Tenure Months"] >= 24
).astype(int)
```
